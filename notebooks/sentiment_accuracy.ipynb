# Sentiment Analysis Accuracy Review

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../static/data/nlp_sentiment_results.csv")
```

```python
df.head()
```

```python
cols = ["listing_id", "vote_sentiment_positivity", "high_conf_sentiment_positivity", 
        "long_conf_sentiment_positivity", "classified_sentiment_positivity",
        "review_scores_rating"
       ]
df = df[cols]
df.head()
```

```python
def true_positivity(row):
    if row["review_scores_rating"] > 80:
        return 1
    else:
        return 0
```

```python
df["true_positivity"] = df.apply(true_positivity, axis=1)
df.head()
```

```python
sum(df["vote_sentiment_positivity"] == df["true_positivity"])/df.shape[0]
```

```python
sum(df["high_conf_sentiment_positivity"] == df["true_positivity"])/df.shape[0]
```

```python
sum(df["long_conf_sentiment_positivity"] == df["true_positivity"])/df.shape[0]
```

```python
sum(df["classified_sentiment_positivity"] == df["true_positivity"])/df.shape[0]
```

According to these accuracy scores, the high confidence measure is the best of them all (although not perfect) and the vote measure is trash.
